# Phương pháp Adaptive Thresholding
Phân đoạn ảnh sử dụng adaptive thresholding để xử lý các vùng có độ sáng khác nhau.

```python
# Phương pháp Adaptive Thresholding

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_local

data = Image.open("exercise/dalat.jpg").convert("L")
a = np.asarray(data)

b = threshold_local(a, 39, offset=10)
plt.imshow(b, cmap='gray')
plt.title('Kết quả Adaptive Thresholding')
plt.axis('off')
plt.show()
```
