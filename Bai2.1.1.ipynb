# Phương pháp Otsu
Phân đoạn ảnh sử dụng phương pháp Otsu để tự động tìm ngưỡng phân tách foreground và background.

```python
# Phương pháp Otsu

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu

data = Image.open("exercise/dalat.jpg").convert("L")
a = np.asarray(data)

thres = threshold_otsu(a)
b = a > thres
plt.imshow(b, cmap='gray')
plt.title('Kết quả Otsu Thresholding')
plt.axis('off')
plt.show()
```
